## Filling area on line plots

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)

data = pd.read_csv('./world_development_indicators/Indicators.csv')
```

### CO2 emission in USA per year

```python
indicator_name = 'CO2 emissions \(metric'    # We need to use a '\' before ( in string

country_code = 'USA'

mask1 = data['IndicatorName'].str.contains(indicator_name)

mask2 = data['CountryCode'].str.contains(country_code)

stage = data[mask1 & mask2]

stage.head()
```

```python
years = stage['Year'].values
Values = stage['Value'].values

plt.plot(years,Values,label='')
plt.axis([1957,2011,0,30])
plt.xlabel('year')
plt.ylabel(stage['IndicatorName'].iloc[0])
plt.title('CO2 emitted by USA')
```

```python
co2='CO2 emissions \(metr'
cc ='CSS'

mask3 = data['IndicatorName'].str.contains(co2)
mask4 = data['CountryCode'].str.contains(cc)

stage1=data[mask3 & mask4]
stage1.head()
```

```python
values1 = stage1['Value'].tolist()

plt.plot(years,values1, label='CSS',linestyle='--', color = 'black')
plt.plot(years,Values,label='USA')
plt.axis([1957,2011,0,30])

plt.xlabel('year')
plt.ylabel(stage['IndicatorName'].iloc[0])
plt.title('CO2 emission')

overall_median = 18 # start filling over the value 10

plt.fill_between(years,Values,overall_median, alpha=0.35) # appha represents the density of color
plt.legend()
plt.show()
```

```python
# fill where the emissions in USA over 18 mertic tons per capita.

#add where=(Values>overall_median),interpolate=True

values1 = stage1['Value'].tolist()

plt.plot(years,values1, label='CSS',linestyle='--', color = 'black')
plt.plot(years,Values,label='USA')
plt.axis([1957,2011,0,30])

plt.xlabel('year')
plt.ylabel(stage['IndicatorName'].iloc[0])
plt.title('CO2 emission')

overall_median = 18 # start filling over the value 10

plt.fill_between(years,Values,overall_median, where=(Values>overall_median),interpolate=True, alpha=0.35)
plt.legend()
plt.show()
```

```python
# fill where the emissions in USA over 18 mertic tons per capita and under 18 metric tons with another color.

values1 = stage1['Value'].tolist()

plt.plot(years,values1, label='CSS',linestyle='--', color = 'black')
plt.plot(years,Values,label='USA')
plt.axis([1957,2011,0,30])

plt.xlabel('year')
plt.ylabel(stage['IndicatorName'].iloc[0])
plt.title('CO2 emission')

overall_median = 18 # start filling over the value 10

plt.fill_between(years,Values,overall_median, where=(Values>overall_median),interpolate=True, alpha=0.35)

plt.fill_between(years,Values,overall_median, where=(Values<=overall_median),interpolate=True,color='red',
                 alpha=0.35)

plt.legend()
plt.show()
```
