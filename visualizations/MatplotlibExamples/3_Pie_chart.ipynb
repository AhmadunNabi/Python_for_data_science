## Pie chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import collections as cn
plt.rcParams.update(plt.rcParamsDefault)

slices = [35,40,25]

labels = ['35%','40%','25%']
colors = ['#454546','#000999','#999999']

plt.pie(slices, labels=labels,colors=colors, wedgeprops={'edgecolor': 'black'})
```

```python
data = pd.read_csv('./world_development_indicators/Indicators.csv')
```

```python
# Counter
indicator = data['IndicatorName'].tolist()

count = cn.Counter(indicator)

m_com_indic = count.most_common(15) # Most 15 counted indicator

m_com_indic

# make chart of counting the most common indicator
indicators = []
counts = []

for incount in m_com_indic:
    indicators.append(incount[0])
    counts.append(incount[1])

print(indicators)
print(counts)

plt.pie(counts, labels = indicators, wedgeprops={'edgecolor':'white'})

plt.show()
```

```python
#it very unwise to use pie plot if we have more than 5 data.
# Counter
indicator = data['IndicatorName'].tolist()

counts=cn.Counter(indicator)

indi = []
num = []

most_count = counts.most_common(5)

for item in most_count:
    indi.append(item[0])
    num.append(item[1])
    
print(indi)
print(num)

explode=[0,0.2,0,0,0]
    
plt.pie(num, labels=indi, explode=explode, shadow=True, startangle = 90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})
plt.show()
```
