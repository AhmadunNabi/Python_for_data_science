## Plotting time series data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dt
from datetime import datetime, timedelta
plt.rcParams.update(plt.rcParamsDefault)

dates=[
    datetime(2020,5,24),
    datetime(2020,5,25),
    datetime(2020,5,26),
    datetime(2020,5,27),
    datetime(2020,5,28),
    datetime(2020,5,29),
    datetime(2020,5,30),
]
y =[0,1,3,4,6,5,7]

plt.plot_date(dates,y)
```

```python
# Solid linestyle
plt.plot_date(dates,y, linestyle='solid')
```

```python
#make the plot little bit smarter
plt.plot_date(dates,y, linestyle='solid')

plt.gcf().autofmt_xdate()                      #gct means get current figure
```

```python
#make the plot little bit smarter
plt.plot_date(dates,y, linestyle='solid')

plt.gcf().autofmt_xdate()                      #gct means get current figure

# Formating date
date_format = dt.DateFormatter('%b,%d %Y')

plt.gca().xaxis.set_major_formatter(date_format)
```

```python
data1 = pd.read_csv('datedata.csv')

data1.head()
```

```python
price_dates = data1['Date']
price_close = data1['Close']

plt.plot_date(price_dates,price_close,linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin prices')
plt.xlabel('Date')
plt.ylabel('closing price')
```

```python
#There 17may is not at the first place. We need sort this sequence. It is necessary to do that thing.
data1['Date'] = pd.to_datetime(data1['Date'])
data1.sort_values('Date', inplace=True)

price_dates = data1['Date']
price_close = data1['Close']

plt.plot_date(price_dates,price_close,linestyle='solid')

plt.gcf().autofmt_xdate()
plt.title('Bitcoin prices')
plt.xlabel('Date')
plt.ylabel('closing price')
plt.show()
```
