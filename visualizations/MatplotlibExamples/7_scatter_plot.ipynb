## Scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
```

```python
x=[5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

plt.scatter(x,y)
```

```python
# change size, color, style
plt.scatter(x,y,s=10, c='green', marker='x')
```

```python
# edge color, linewidth
plt.scatter(x,y,s=40, c='green', edgecolor='black',
            linewidth=0.5, alpha=.95) # Alpha means density of color.
```

```python
# Suppose we had a servey of satisfaction label between 0-10. We can represent that rating according to color.
# edge color, linewidth
color = [9,6,7,5,7,8,4,5,1,3,2,1,5,7,8,5,6,7,9,8]
plt.scatter(x,y,s=100, c=color, cmap='Greens', edgecolor='black',
            linewidth=0.5, alpha=1) # Alpha means density of color.

cbar=plt.colorbar()
cbar.set_label('satisfaction')
```

```python
# we can change size of data
size=[45,30,35,25,35,40,20,25,5,15,10,5,25,35,40,25,30,35,45,40]

plt.scatter(x,y, c=color, s=size, cmap='Greens', edgecolor='black',
            linewidth=0.5, alpha=1) # Alpha means density of color.

cbar=plt.colorbar()
cbar.set_label('satisfaction')
```

```python
youtubedata=pd.read_csv('youtubedata.csv')
youtubedata.head()
```

```python
view_count=youtubedata['view_count']

likes=youtubedata['likes']

ratio=youtubedata['ratio']

plt.scatter(view_count,likes, edgecolor='black',linewidth=0.5,alpha=0.75)
```
