## Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
```

```python
ages =[18,19,21,25,26,26,30,32,38,45,55]
plt.hist(ages,bins=5,edgecolor='black')
plt.show()
```

```python
# We can set the bins as a list of data
ages =[18,19,21,25,26,26,30,32,38,45,55]
bins =[10,20,30,40,55]
plt.hist(ages,bins=bins,edgecolor='black')
plt.show()
```

### From a dataset

```python
data = pd.read_csv('./world_development_indicators/Indicators.csv')
```

```python
indicator_name = 'CO2 emissions \(metric'    # We need to use a '\' before ( in string

country_code = 'USA'

mask1 = data['IndicatorName'].str.contains(indicator_name)

mask2 = data['CountryCode'].str.contains(country_code)

stage = data[mask1 & mask2]
```

```python
# From a dataset 
# Find the range of co2 emission from USA in maximum year
stage.head()
```

```python
hist_data = stage['Value'].values

print(hist_data.min())
print(hist_data.max())
```

```python
plt.hist(hist_data, bins=15,edgecolor='black')
```

```python
bins = [15,15.8,16.6,17.4,18.2,19,19.8,20.6,21.4,22.2]
plt.hist(hist_data,density=False, bins=bins,edgecolor='black') # If you don't use density=False no problem
```
