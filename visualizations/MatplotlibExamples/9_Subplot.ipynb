## Subplot
If we want additional plot or if we want to work with plots in a more object oriented manner then it's best to create our plots in subplot method.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
#plt.style.use('fivethirtyeight')


s_data=pd.read_csv('s_data.csv')

ages = s_data['Age']
all_devs = s_data['All_Devs']
python = s_data['Python']
js = s_data['JavaScript']

plt.plot(ages,all_devs, label='All devs')
plt.plot(ages,python, label='Python')
plt.plot(ages,js,color='#444444', linestyle='--',label='JavaScript')

plt.legend()
s_data.head()

plt.title('Median Salary(USD) by age')
plt.xlabel('age')
plt.ylabel('USD')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

s_data=pd.read_csv('s_data.csv')

ages = s_data['Age']
all_devs = s_data['All_Devs']
python = s_data['Python']
js = s_data['JavaScript']
fig, (ax1,ax2)=plt.subplots(nrows=2,ncols=1, sharex=True) # two rows one column 
#share x to use the mark in only one axis

ax1.plot(ages,all_devs, label='All devs')
ax2.plot(ages,python, label='Python')
ax2.plot(ages,js,color='#444444', linestyle='--',label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary(USD) by age')

ax1.set_ylabel('USD')

ax2.legend()
ax2.set_xlabel('age')
ax2.set_ylabel('USD')
ax1.grid()
ax2.grid()
plt.show()
```

Here we made two plot in one figure.
We can eaxity make two plot in different two figure

```python
import pandas as pd
import matplotlib.pyplot as plt

s_data=pd.read_csv('s_data.csv')

ages = s_data['Age']
all_devs = s_data['All_Devs']
python = s_data['Python']
js = s_data['JavaScript']

# Here we need to make a change
fig1, ax1=plt.subplots()  
fig2, ax2=plt.subplots() 

ax1.plot(ages,all_devs, label='All devs')
ax2.plot(ages,python, label='Python')
ax2.plot(ages,js,color='#444444', linestyle='--',label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary(USD) by age')
ax1.set_xlabel('age')
ax1.set_ylabel('USD')

ax2.legend()
ax2.set_title('Median Salary(USD) by age')
ax2.set_xlabel('age')
ax2.set_ylabel('USD')
ax1.grid()
ax2.grid()
plt.show()

#we can save the figure
fig1.savefig('fig1.png')
fig2.savefig('fig1.png')
```
