## Frequent string opration

```python
import pandas as pd

movies = pd.read_csv('./movielens/movies.csv')

movies.head()
```

### split function

```python
movie_genres = movies['genres'].str.split('|', expand = True)

movie_genres
```

Normally split operation returns a series of lists, by providing expand = True operation we make sure the output is not just a series of list it's an actual dataframe.

```python
# Add another column which says the genres is comedy, True or False

movie_genres['is_comedy'] = movies['genres'].str.contains('Comedy')

movie_genres
```

### Extract year from title

```python
movies['year'] = movies['title'].str.extract('.*\((.*)\).*', expand = True)

movies['year']
```

```python
ratings = pd.read_csv('./movielens/ratings.csv')

total = movies.merge(ratings, on = ('movieId'), how = 'inner')

latest = total['year'] > '2011'

highly_rated = total['rating'] == 5.0

romantic = total['genres'].str.contains('Romance')

total[romantic]

latest_highly_rated_romantic = total[latest & highly_rated & romantic]

latest_highly_rated_romantic #= latest_highly_rated_romantic.groupby('movieId')

latest_highly_rated_romantic
```

```python
latest_highly_rated_romantic.drop_duplicates('movieId', keep='last')
```
