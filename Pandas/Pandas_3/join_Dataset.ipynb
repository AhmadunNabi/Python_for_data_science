```python
import pandas as pd

movies = pd.read_csv('./movielens/movies.csv')

movies.head()
```

```python
tags = pd.read_csv('./movielens/tags.csv')

del tags['timestamp']

tags.head()
```

```python
ratings = pd.read_csv('./movielens/ratings.csv')

del ratings['timestamp']

ratings.head()
```

# Merge Datasets

```python
t = movies.merge(tags, on ='movieId', how = 'inner')

t.head()
```

```python
f = t.merge(ratings, on =('movieId','userId'), how = 'inner')

f.head
```

```python
left = pd.DataFrame({'key1': ['k0','k0','k1','k2'],
                     'key2': ['k0','k1','k0','k1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key1': ['k0','k1','k1','k2'],
                     'key2': ['k0','k0','k0','k0'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']})

result = pd.merge(left,right, how = 'left', on = ['key1','key2'])

result
```

```python
result = pd.merge(left, right, how='outer', on=['key1', 'key2'])

result
```

```python
result = pd.merge(left, right, how='inner', on=['key1', 'key2']) # Defult is inner

result
```

```python
result = pd.merge(left, right, how='right', on=['key1', 'key2'])

result
```

## concat() function

```python
# Stack Dataframe

pd.concat([movies,ratings])
```

```python
pd.concat([movies,ratings], keys = ['movies','ratings'])
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
print('Array:\n',df1)

frames = [df1,df2,df3]

print('\nArrays:\n',frames)

result = pd.concat(frames)

result
```

```python
result = pd.concat(frames, keys=['1st', '2nd', '3rd'])

result
```

**we can now select out each chunk by key**

```python
result.loc['3rd']
```

```python
# join = 'outer' means union of the dataframes

# join = 'inner' means intersection of the dataframes.
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
result = pd.concat([df1,df4], axis = 1, join = 'inner')  # axis = 1 means compare with rows.

result                      # without axis means defalt axis = 0 means compare between columns.
```

**Concating using append**

## Combine aggregation, merging and filters to get an useful analytics

```python
avg_rating = ratings.groupby('movieId', as_index = False).mean()

del avg_rating['userId']

avg_rating.head()
```

```python
box_office = movies.merge(avg_rating, on = 'movieId', how = 'inner')

box_office.tail()
```

```python
is_highly_rated = box_office['rating'] >= 4.0

box_office[is_highly_rated]
```

```python
#Find highly rated comedy movies

is_comedy = box_office['genres'].str.contains('Comedy')

box_office[is_comedy]

box_office[is_comedy & is_highly_rated]
```
