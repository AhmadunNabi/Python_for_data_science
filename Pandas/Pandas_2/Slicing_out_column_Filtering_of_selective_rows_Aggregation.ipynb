## Aggregation

Aggregating values across rows gives us a big picture of a dataset

```python
ratings = ratings.fillna(ratings['rating'].max())

ratings_count = ratings[['movieId','rating']].groupby('rating').count()

ratings_count
```

```python
average_rating = ratings[['movieId','rating']].groupby('movieId').mean()

average_rating[-10:]
```

```python
movie_count = ratings[['movieId','rating']].groupby('movieId').count()

print(movie_count)

movie_count[:10].plot.bar(figsize = (15,10))
```
